```python
import glob
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
```

```python
# Laden des Datensatzes
cur_path = "C:/Users/de136581/Documents/Uni/SoSe_2024/DSKI_Projekt/Data_trafic_Signs/GTSRB/"  
data = []
labels = []
classes = 43
```

Read ppm Datein into python

```python
for i in range(classes):
    path = cur_path + 'Training/' + str(i).zfill(5) + '/*'
    images = glob.glob(path)
    for im in images:
        try:
            image = cv2.imread(im)
            image = cv2.resize(image, (30, 30))
            data.append(image)
            labels.append(i)
        except:
            print("Fehler beim Laden des Bildes")

data = np.array(data)
labels = np.array(labels)
```

```python
print(data)
```

```python
# Train-Test-Aufteilung
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# One-Hot-Encoding der Labels
y_train = to_categorical(y_train, 43)
y_test = to_categorical(y_test, 43)
```

```python
# Modellarchitektur
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu', input_shape=X_train.shape[1:]))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Genauigkeit: {test_acc:.4f}")
```
